```python
# loading the required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# load the dataset on and store it in object fifa
fifa=pd.read_csv("fifa data.csv")
fifa
```

```python
## to get all column names
for i in fifa.columns:
    print(i)
```

```python
## to check the number of rows and columns 
fifa.shape
```

```python
# to get nationality and there count 
fifa['nationality'].value_counts()
```

```python
## to get the top five nations which have higest number of counts
fifa['nationality'].value_counts()[:5].keys()
```

```python
##bar graph between top 5 nations and there counts 
plt.figure()
plt.bar(list(fifa['nationality'].value_counts()[:5].keys()),list(fifa['nationality'].value_counts()[:5]))
plt.show()
```

```python
## to check who gets higest salaries()
fifa[["short_name","wage_eur"]]
```

```python
player_salary=fifa[["short_name","wage_eur"]].sort_values(by="wage_eur",ascending=False)[0:5]
player_salary
```

```python
plt.figure(figsize=(8,5)) ## figure size is defined
plt.bar(list(player_salary['short_name']),list(player_salary["wage_eur"]),color=("red","blue","pink","yellow","orange"))
plt.show()
```

```python
## to find football players who has more weight,height & wages in germany
fifa['nationality']=="germany"
```

```python
germany=fifa[fifa["nationality"]=="Germany"]
germany.head()
```

```python
# to find the tallest german player
b=germany.sort_values(by=["height_cm"],ascending=False)
b["short_name"].head(1)
```

```python
# to find the german player who has more weight
a=germany.sort_values(by=["weight_kg"],ascending=False)
a["short_name"].head(1)
```

```python
# to find the top 5 german players who earns more salaries then others 
c=germany.sort_values(by=["wage_eur"],ascending=False)
c[["short_name",'wage_eur']].head(5)
```

```python
# to find the player who is best in shooting
k=fifa[["short_name","shooting"]]
k.sort_values(by=['shooting'],ascending=False).head(5)
```

```python
## to find top five defenders,their nationality and their club
player_defending=fifa[["short_name","defending",'nationality','club']]
player_defending.sort_values(by=['defending'],ascending=False).head(5)
```

```python
#to find the higest earning players in real madrid
real_madrid=fifa[fifa['club']=='Real Madrid']
real_madrid.sort_values(by=['wage_eur'],ascending=False).head()
```
